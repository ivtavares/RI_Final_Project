# Download Files from Github

```python
from selenium import webdriver
import urllib.request
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait # available since 2.4.0
from selenium.webdriver.support import expected_conditions as EC # available since 2.26.0
```

```python
# Prepare the code for easily opening new browser window
driver_option = webdriver.ChromeOptions()
driver_option.add_argument(" — incognito")
chromedriver_path = '/home/igor/Estudos/Ciencia de Dados e Big Data/RI/chromedriver_linux'

def create_webdriver():
    return webdriver.Chrome(executable_path=chromedriver_path, options = driver_option)
```

```python
#Leitura e modificaçao do arquivo
import pandas as pd
list_pd = pd.read_csv('/home/igor/Estudos/Ciencia de Dados e Big Data/RI/project_list.csv')
list_pd['project_name'].replace(to_replace=' / ', value = '-', regex = True, inplace=True)
list_pd
```

```python
def download_github(list_url):
    '''Function to download eat repository from
    github link list_url'''
    for i in range(len(list_url)-1):
        url = list_url.iloc[i,0]
        name = list_url.iloc[i,1]
        print(url)
        browser = create_webdriver()
        browser.get(url)
        zip_file = browser.find_elements_by_xpath('//div[*]/details/summary')[0].text[8:]
        print('Beginning file download ' + name + ' with urllib2...')
        url_download = url + "/archive/"+ zip_file +'.zip'  
        urllib.request.urlretrieve(url_download, 
                                   '/home/igor/Estudos/Ciencia de Dados e Big Data/RI/Github' + name + '.zip')
        browser.quit()
```

```python
download_github(list_pd)
```
